```python
def f(x, y, z): # we call x y z "parameters"
    result = x + y + z
    return result

f(2, 4, 6) # we call 2 4 6 "arguments"

def f_2(x, y, z=7): # we call z=7 "default value"
    result = x + y + z
    return result
f_2(2, 4, 6)
f_2(2, 4) # when you have default value, you

"""
but you cannot set default value (x=2, y, z=7)
you cannot put the default before the parameter
"""

print(f(2, 3, z=1))
```
