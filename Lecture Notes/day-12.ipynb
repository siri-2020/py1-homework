```python
empty_dict = {}
empty_dict = dict()

animal_legs = {'human':2, 'horse':4, 'fish':0, 'cheetah':4}
animal_legs['cheetah']
animal_legs.get('fish')
print(animal_legs.get('cat'))
animal_legs.get('cheetahs', 'do not found the key') # 3 is a fall back if dict don't have that key
```

```python
# loop by key
for key in animal_legs:
    print(key)
    print(animal_legs[key])

# loop by .item()
for item in animal_legs.items():
    print(item, type(item))
for key, value in animal_legs.items():
    print(key)
    print(value)

for key in animal_legs.keys():
    print(key)
for value in animal_legs.values():
    print(value)
```
